```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
```

```python
# Define State

class BMIState(TypedDict, total=False):
    height_m: float
    weight_kg: float
    bmi: float
    bmi_category: str
```

```python
def calculate_bmi(state: BMIState) -> BMIState:
    weight_kg = state["weight_kg"]
    height_m = state["height_m"]
    bmi = weight_kg / (height_m ** 2)
    state["bmi"] = bmi
    
    # Basic BMI category assignment
    if bmi < 18.5:
        state["bmi_category"] = "Underweight"
    elif bmi < 25:
        state["bmi_category"] = "Normal weight"
    elif bmi < 30:
        state["bmi_category"] = "Overweight"
    else:
        state["bmi_category"] = "Obese"
    
    return state
```

```python
#Define graph

graph = StateGraph(BMIState)

# add nodes to graph

graph.add_node("calculate_bmi", calculate_bmi)


# add edges to graph

graph.add_edge(START, "calculate_bmi")



# compile the graph

workflow = graph.compile()



```

```python
# excute the graph

intial_state = {'weight_kg': 70, 'height_m': 1.75}

final_state = workflow.invoke(intial_state)

print(final_state)

```
